```python
path = kagglehub.dataset_download("lava18/google-play-store-apps")
print("Path to dataset files:", path)
```

```python
data=pd.read_csv(path+"/googleplaystore.csv")
print("Dataset Loaded Properly.")
```

```python
data.isnull().sum()
```

```python
data['Rating'].fillna(data['Rating'].mean(),inplace=True)
```

```python
data.drop_duplicates(inplace=True)
print("Duplicate Rows removed , newshape is :",data.shape)
```

```python
data['Price']=data['Price'].astype(str).str.strip()
data['Installs']=data['Installs'].astype(str).str.strip()
data['Reviews']=data['Reviews'].astype(str).str.strip()


data['Price']=data['Price'].str.replace('$','',regex=False)
data['Price']=pd.to_numeric(data['Price'],errors='coerce')


data['Installs']=data['Installs'].str.replace('+','',regex=False)
data['Installs']=data['Installs'].str.replace(',','',regex=False)
data['Installs']=pd.to_numeric(data['Installs'],errors='coerce')

data['Reviews']=pd.to_numeric(data['Reviews'],errors='coerce')

data.dropna(subset=['Price','Installs','Reviews'],inplace=True)

print("data Type Cleaned Done")
data[['App','Price','Installs','Reviews']].head()
```

```python
data=data[data['Rating']<=5]
```

```python
data['isPaid']=data['Type'].apply(lambda x:1 if x=='Paid' else 0)
data[['App','Type','isPaid']].head()
```

```python
data.to_csv('cleaned_playstore.csv',index=False)
print("Cleaned Dataset saved as cleaned_playstore.csv")
```

### Distribution of App Sizes

Let's examine the distribution of app sizes to understand how they vary across the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Clean the 'Size' column and convert to numeric
data['Size'] = data['Size'].astype(str).str.replace('M', 'e6').str.replace('k', 'e3')
data['Size'] = pd.to_numeric(data['Size'], errors='coerce')
data.dropna(subset=['Size'], inplace=True)


# Visualize the distribution of app sizes
plt.figure(figsize=(10, 6))
sns.histplot(data['Size'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of App Sizes')
plt.xlabel('Size')
plt.ylabel('Frequency')
plt.xticks(rotation=90) # Rotate x-axis labels for better readability
plt.show()
```

```python
top_categories=data['Category'].value_counts().head(10)
print("Top 10 Categories:\n",top_categories)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values,y=top_categories.index,palette='viridis', legend=False)
plt.title("Top 10 App Categories")
plt.xlabel("Number of Apps")
plt.ylabel("Category")
plt.show()
```

```python
avg_rating=data.groupby('Category')['Rating'].mean().sort_values(ascending=False).head(10)
print("Average Ratings by Category:\n",avg_rating)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating.values,y=avg_rating.index,palette='viridis')
plt.title("Average Ratings by Category")
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

```python
app_types=data['Type'].value_counts()
print("App Types:\n",app_types)
plt.figure(figsize=(8,6))
sns.barplot(x=app_types.index,y=app_types.values,palette='coolwarm')
plt.title("Free vs paid app distribution")
plt.xlabel("Number of Apps")
plt.ylabel("Type")
plt.show()
```

```python
 plt.figure(figsize=(10,6))
 sns.histplot(data['Rating'],bins=20,kde=True,color='darkred')
 plt.title("Distribution of App Prices")
 plt.xlabel("Rating")
 plt.ylabel("Frequency")
 plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Price',y='Rating',data=data,alpha=0.6,color='orange' )
plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Installs',y='Rating',data=data,alpha=0.6,color='green',hue='Type')
plt.xscale('log')
plt.yscale('log')
plt.title("Installs vs Rating")
plt.xlabel("Rating")
plt.ylabel("Installs")
plt.legend(title='AppZType')


plt.show()
```

```python
corr=data[['Rating','Reviews','Installs','Price','isPaid']].corr()
print("Correlation Matrix:\n",corr)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidths=0.5,fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```
